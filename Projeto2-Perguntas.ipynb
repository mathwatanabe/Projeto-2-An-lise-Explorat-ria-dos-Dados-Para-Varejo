## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

<b>Para responder a primeira pergunta proposta, organizou-se o dataframe pelos campos que apresentam a categoria "Office Supplies".</b>

```python
# Buscando na coluna "Categoria" apenas a correspondência "Office Supplies"
d0 = df_dsa[df_dsa['Categoria'] == 'Office Supplies']
d0
```

<b>Para apresentar os valores totais para cada cidade, foi realizado um <code>groupby</code> que organiza uma coluna específica ou colunas específicas agrupadas. Como foi pedido o valor máximo de vendas, cada cidade foi organizada pela soma total dos valores encontrados na coluna "Valor_Venda". Para organizar o dataframe em ordem decrescente e assim apresentar, na primeira linha, o maior valor de vendas presente foi utilizada a função <code>sort_values</code>, substituindo o valor do atributo <code>ascending</code> por <code>False</code>.</b>

```python
d1 = d0.groupby('Cidade')['Valor_Venda'].sum().reset_index()
d1.sort_values(by='Valor_Venda',ascending=False)
```

<b>Por último, para dar a resposta da pergunta apresentada anteriormente, foi utilizada a função <code>loc</code> para retornar os dados de uma linha de um dado índice. Como o desejado foi o valor máximo, a função <code>idxmax()</code> retorna o índice do valor máximo em um dataframe.</b>

```python
d1_max_ValorVenda = d1.loc[d1['Valor_Venda'].idxmax()]
print(d1_max_ValorVenda,'\n')
a = d1_max_ValorVenda['Cidade']
b = d1_max_ValorVenda['Valor_Venda']
print(f'O maior valor de venda, na categoria Office Supplies, ocorreu na cidade de {a} com o respectivo valor de U$ {b}.')
```

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

<b>O total de vendas por cada data foi realizado abaixo. Foram realizadas duas formas, uma organizando de forma crescente as datas e outra em como o dataframe apresentou os valores, sem uma organização prévia.</b>

```python
#Agrupando por cada data e apresentando o valor total das vendas de cada data
d2_1 = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()
d2_1_indexreset = d2_1.reset_index()
d2_1_indexreset
```

<b>Como as datas foram organizadas pela ordem dia/mês/ano, foi necessário o tratamento desses dados, pois, nessa organização, como a coluna apresenta dados do tipo string, organizá-lo de forma crescente iria considerar os caracteres na ordem que foram apresentados. Ou seja, todos os dias 01 viriam antes, independente do respectivo ano. Poderia ser utilizado a biblioteca datatime para transformar o tipo de dados dessa coluna para o tipo data. No entanto, optou-se por utilizar outro caminho mais trabalhoso para treinar alguns fundamentos da linguagem Python.</b>

```python
#Guardando o dataframe por Data com a soma de cada valor de venda para cada data
d2_1 = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()
d2_1_indexreset = d2_1.reset_index()
data_invertida = []
data_primeira = []
#Quebrando cada valor dos campos de data por "/"
for i in d2_1_indexreset['Data_Pedido'].str.split('/'):
    #Alterando a ordem para ano,mês,dia
    data_primeira.append(i[2])
    data_primeira.append(i[1])
    data_primeira.append(i[0])
    #Adicionando a data invertida na lista que será utilizada para construir a nova coluna
    data_invertida.append(data_primeira)
    #Limpando a lista para guardar a próxima data invertida
    data_primeira = []
data_com_barra = []
#Formatando os valores para apresentarem as datas com "/"
for i in data_invertida:
    p = i[0] + '/' + i[1] + '/' + i[2]
    data_com_barra.append(p)
d2_1_indexreset['Data_Invertida'] = data_com_barra
#Organizando pela primeira data até a última
d2_organizadas = d2_1_indexreset[['Data_Invertida','Valor_Venda']].sort_values(by='Data_Invertida')
print(d2_organizadas)
#Apresentando a última data e o maior valor de vendas encontrado
print(d2_organizadas.loc[d2_organizadas.idxmax()])
```

```python
#Definindo os dados dos eixos
x = d2_organizadas['Data_Invertida']
y = d2_organizadas['Valor_Venda']
#Definindo o tamanho da imagem do gráfico
fig, axe = plt.subplots(figsize=(20,8))
axe.plot(x,y)
#Definindo a escala apresentada do eixo x e os valores apresentados
plt.setp(axe, xticks=['2015/03/18','2015/10/20','2016/08/07','2017/05/26','2018/03/14','2018/12/31'], xticklabels=['18/03/2015','20/10/2015','07/08/2016','26/05/2017','14/03/2018','31/12/2018'])
plt.xlabel('Data do Pedido')
plt.ylabel('Valor das Vendas')
plt.title('Valor de Vendas por Data')
plt.show()
```

```python
#Apresentando o maior valor apresentado no gráfico
data_da_maior_venda = d2_1.idxmax()
valor_maior_venda = d2_1.max()
print(f'A data que ocorreu o maior valor total de vendas foi {data_da_maior_venda} e o valor foi de {valor_maior_venda}.')
```

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

<b>Para obter o segmento com o maior valor total de vendas, agrupou-se o dataframe por segmento. Havia três segmentos possíveis "Consumer", "Corporate" e "Home Office".</b>

```python
d5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index()
d5
```

<b>O segmento com o maior valor de vendas é apresentado abaixo e foi obtido pela função <code>loc</code>.</b>

<b>Como os valores apresentados anteriormente estão em notação científica, ou seja, estão em potência de 10, eles precisam ser transformados em valores extensos. A função criada abaixo realiza tão processo para a posterior plotagem dos dados em um gráfico de pizza.</b>

```python
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
colors5 = sns.color_palette('pastel')[0:5]
#Destacando a fatia Corporate
explode = (0,0.1,0.02)
plt.pie(d5['Valor_Venda'],
        explode=explode,
        labels = d5['Segmento'],
        colors = colors5,
        autopct='%.2f%%',
        shadow = True)
plt.show()
```

<b>O segmento que contribuiu em maior proporção com as vendas, entre os anos avaliados, foi o "Consumer". Apresentando os dados de uma forma diferente, temos abaixo.</b>

```python
plt.pie(d5['Valor_Venda'],
        labels = d5['Segmento'],
        autopct = autopct_format(d5['Valor_Venda']),
        colors = colors5)
# Transforma o gráfico pie em anel
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Labels e anotações
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(d5['Valor_Venda']))), xy = (-0.65, -0.05))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

<b>Foram realizadas várias formas de responder a essa pergunta. Posteriormente, foi realizada a contagem de quantos valores essa consulta retornou, pois cada valor retornada equivaleria a uma venda que receberia o desconto de 15%.</b>

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

<b>Para calcular o valor da média de venda antes e depois do desconto de 15%, foi construído o dataframe com os novos valores de venda, aplicando os respectivos descontos. Foi utilizada a função <code>where()</code> para buscar onde houvesse o valor de 0.1, ou seja, 10% de desconto, a nova coluna "Valor_Venda_com_Desconto" apresentaria o valor de 90% do valor antes do desconto. Analogamente, 85% para os campos que não apresentassem o 0.1, pois esses receberiam o desconto de 15%, ou seja, o novo valor de venda corresponderia a 85%.</b>

```python
df_dsa['Valor_Venda_com_Desconto'] = np.where(df_dsa['Desconto'] == 0.1, df_dsa['Valor_Venda']*0.9, df_dsa['Valor_Venda']*0.85)
df_dsa
```

<b>Foram utilizados 4 novas visualizações do dataframe. 2 antes dos descontos e 2 depois dos descontos. Essa lógica foi utilizada para calcular a média de cada um deles.</b>

```python
#Plotagem para a mediana
plt.figure(figsize=(12,6))
figura1 = sns.relplot(kind='line',
                    data=d9,
                    y = 'median',
                    x = meses,
                    hue = segmentos,
                    col = anos,
                    col_wrap = 4)
plt.show()
```

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

<b>Para realizar a organização por categoria e por subcategoria, foi utilizada a linha de código abaixo. O <code>groupby</code> foi utilizado para agrupar as colunas por "Categoria" e por "Subcategoria", o <code>sum()</code> foi utilizado para somar o valor de venda. Para arredondar os valores, a função <code>astype()</code> foi utilizada para converter os valores float para int. O <code>sort_values()</code> foi utilizado para organizar de decrescente os valores de venda e, por último, para apresentar os 12 maiores, foi utilizado o <code>head()</code> com o parâmetro 12.</b>

```python
#Pode usar .sum(numeric_only = True) para realizar a soma apenas para os numéricos
d10 = df_dsa.groupby(['Categoria','SubCategoria'])['Valor_Venda'].sum().astype(int).sort_values(ascending=False).reset_index().head(12)
d10
```

```python
# Organizando por Categoria para ter uma melhor visualização
d10 = d10.sort_values(by='Categoria')
d10
```

```python
# Somando os valores de venda para cada Categoria
d10_cat = d10.groupby(['Categoria'])['Valor_Venda'].sum().reset_index()
d10_cat
```

```python
# Definindo as cores para as categorias
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
# Definindo as cores para as subcategorias
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
fig, ax = plt.subplots(figsize = (18,12))
#Criando um gráfico mais externo para as categorias
p1 = ax.pie(d10_cat['Valor_Venda'],
            radius = 1,
            labels = d10_cat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)
#Criando um gráfico mais interno para as subcategorias
p2 = ax.pie(d10['Valor_Venda'],
            radius = 0.9,
            labels = d10['SubCategoria'],
            autopct = autopct_format(d10['Valor_Venda']),
            colors = cores_subcategorias,
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'),
            pctdistance = 0.53,
            rotatelabels = True)

#Colocando o centro do gráfico branco
centre_circle = plt.Circle((0,0), 0.6, fc = 'white')

fig = plt.gcf()
fig.gca().add_artist(centre_circle)
#Valor total de vendas no centro
plt.annotate(text = 'Total de Vendas: ' + '$' + str(int(sum(d10['Valor_Venda']))), xy = (-0.2,0))
plt.title('Total de Vendas por Categoria e 12 Subcategorias mais vendidas')
plt.show()
```
